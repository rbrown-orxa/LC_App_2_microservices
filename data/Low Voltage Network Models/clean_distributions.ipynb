```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

## connection times, this is when the EVs are connected to the charging point

## Web plot digitizer conversion
https://apps.automeris.io/wpd/

```python
conn_times = pd.read_csv('connection_times.csv', header=None, index_col=[0]).rename(columns={1:'probability'})
```

```python
periods = range(1, 24*60//15)
periods = pd.DataFrame([np.nan]*len(periods), index=periods, columns=['null'])
```

```python
conn_times = periods.merge(conn_times, left_index=True, right_index=True, how='outer').drop(columns='null')
```

```python
conn_times.index = pd.to_timedelta( (conn_times.index -1) * 15, unit='min' )
```

```python
n = 8
ax = conn_times.plot(kind='bar', figsize=(8,5),
    title='../cleaned/low_voltage_network_solutions_connection_times.csv')
ticks = ax.xaxis.get_ticklocs()
ticklabels = [str(l.get_text()).split()[-1] for l in ax.xaxis.get_ticklabels()]
z = ax.xaxis.set_ticks(ticks[::n])
z = ax.xaxis.set_ticklabels(ticklabels[::n])
```

```python
conn_times.to_csv('../cleaned/low_voltage_network_solutions_connection_times.csv')
```

```python
conn_energies = pd.read_csv('connection_energy.csv', header=None, index_col=[0]).rename(columns={1:'probability'})
```
